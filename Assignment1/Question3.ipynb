```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load grayscale image
img = cv2.imread("Image1_gray.jpg", cv2.IMREAD_GRAYSCALE)
#img = img.astype(np.float64)

plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.show()
```

```python
mask = 255 - img

plt.figure(figsize=(6,6))
plt.imshow(mask, cmap='gray')
plt.axis("off")
plt.show()
```

```python
alpha = [-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75]
i = 0
plt.figure(figsize=(10,6))
for k in alpha:
    corrected = img + k*mask
    corrected = np.clip(corrected, 0, 255)
    
    i = i+1
    
    plt.subplot(2, 4, i)
    plt.imshow(corrected, cmap='gray')
    plt.axis('off')
    plt.title(f"alpha = {k}")
    
plt.tight_layout()
plt.show()

corrected = img + 0.5*mask
corrected = np.clip(corrected, 0, 255)
```

```python
# Define weights (alpha values)
weights = np.array([0.3, 0.5, 0.7, 0.9, 1, 1.3, 1.5])

results = []
i = 2

plt.figure(figsize=(10, 6))
plt.subplot(2, 4, 1)
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.title(f"original")

for alpha in weights:
    out = img + alpha * corrected          # weighted addition
    out = np.clip(out, 0, 255)
    out = out.astype(np.uint8)

    plt.subplot(2, 4, i)
    plt.imshow(out, cmap='gray')
    plt.axis("off")
    plt.title(f"beta = {alpha}")

    results.append(out)
    i += 1

plt.tight_layout()
plt.show()
```
